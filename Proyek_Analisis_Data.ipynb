## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
# Membaca dataset
df_hour = pd.read_csv('hour.csv')
df_day = pd.read_csv('day.csv')
```

**Insight:**
- Membaca dua file csv yaitu hour.csv dan day.csv
- Dataset telah berhasil dibaca dan siap untuk dianalisis.

### Assessing Data

```python
# Melihat struktur dataset
print(df_hour.head())
print(df_day.head())

# Melihat informasi tentang dataset
print(df_hour.info())
print(df_day.info())

# Melihat statistik deskriptif tentang dataset
print(df_hour.describe())
print(df_day.describe())
```

**Insight:**
- Dataset "hour.csv" memiliki 17 kolom dan 17379 baris
- Dataset "day.csv" memiliki 16 kolom dan 731 baris

### Cleaning Data

```python
# Menghapus baris dengan nilai kosong
df_hour.dropna(inplace=True)
df_day.dropna(inplace=True)

# Mengubah tipe data kolom 'dteday' menjadi datetime
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
```

**Insight:**
- Dataset telah berhasil dibersihkan dan siap untuk dianalisis.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Melihat distribusi data
sns.set()
plt.figure(figsize=(10,6))
sns.distplot(df_hour['cnt'])
plt.title('Distribusi Penggunaan Sepeda Berbagi (Jam)')
plt.xlabel('Jumlah Penggunaan Sepeda Berbagi')
plt.ylabel('Frekuensi')
plt.show()

sns.set()
plt.figure(figsize=(10,6))
sns.distplot(df_day['cnt'])
plt.title('Distribusi Penggunaan Sepeda Berbagi (Hari)')
plt.xlabel('Jumlah Penggunaan Sepeda Berbagi')
plt.ylabel('Frekuensi')
plt.show()
```

**Insight:**
- Distribusi data menunjukkan bahwa penggunaan sepeda berbagi memiliki pola yang tidak terlalu berbeda antara hari kerja dan hari libur.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi data 1: Pengaruh cuaca terhadap penggunaan sepeda berbagi (jam)
sns.set()
plt.figure(figsize=(10,6))
sns.barplot(x='weathersit', y='cnt', data=df_hour)
plt.title('Pengaruh Cuaca terhadap Penggunaan Sepeda Berbagi (Jam)')
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Penggunaan Sepeda Berbagi')
plt.show()

# Visualisasi data 1: Pengaruh cuaca terhadap penggunaan sepeda berbagi (hari)
sns.set()
plt.figure(figsize=(10,6))
sns.barplot(x='weathersit', y='cnt', data=df_day)
plt.title('Pengaruh Cuaca terhadap Penggunaan Sepeda Berbagi (Hari)')
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Penggunaan Sepeda Berbagi')
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi data 2: Perbedaan penggunaan sepeda berbagi antara hari kerja dan hari libur (jam)
sns.set()
plt.figure(figsize=(10,6))
sns.barplot(x='weekday', y='cnt', data=df_hour)
plt.title('Perbedaan Penggunaan Sepeda Berbagi antara Hari Kerja dan Hari Libur (Jam)')
plt.xlabel('Hari')
plt.ylabel('Jumlah Penggunaan Sepeda Berbagi')
plt.show()

# Visualisasi data 2: Perbedaan penggunaan sepeda berbagi antara hari kerja dan hari libur (hari)
sns.set()
plt.figure(figsize=(10,6))
sns.barplot(x='weekday', y='cnt', data=df_day)
plt.title('Perbedaan Penggunaan Sepeda Berbagi antara Hari Kerja dan Hari Libur (Hari)')
plt.xlabel(' Hari')
plt.ylabel('Jumlah Penggunaan Sepeda Berbagi')
plt.show()
```

**Insight:**
- Visualisasi data menunjukkan bahwa cuaca memiliki pengaruh yang signifikan terhadap penggunaan sepeda berbagi, dan bahwa ada perbedaan penggunaan sepeda berbagi antara hari kerja dan hari libur.

###Pertanyaan 3:

```python
# Visualisasi data 3: Persentase penggunaan sepeda pada hari kerja dan hari libur (jam)
plt.figure(figsize=(10,6))
labels = df_hour['weekday'].value_counts().index
plt.pie(df_hour['weekday'].value_counts(), labels=labels, autopct='%1.1f%%')
plt.title('Persentase Penggunaan Sepeda pada Hari Kerja dan Hari Libur (Jam)')
plt.show()

# Visualisasi data 3: Persentase penggunaan sepeda pada hari kerja dan hari libur (hari)
plt.figure(figsize=(10,6))
labels = df_day['weekday'].value_counts().index
plt.pie(df_day['weekday'].value_counts(), labels=labels, autopct='%1.1f%%')
plt.title('Persentase Penggunaan Sepeda pada Hari Kerja dan Hari Libur (Hari)')
plt.show()
```

**Insight:**
- persentase penggunaan sepeda pada hari kerja dan hari libur yang terbagi rata antara 7 kategori (0 sampai 6), dengan masing-masing kategori memiliki persentase penggunaan yang hampir sama (sekitar 14%).

###Pertanyaan 4:

```python
# Visualisasi data 4: Persentase rata-rata jumlah pengguna sepeda berdasarkan musim (jam)
plt.figure(figsize=(10,6))
plt.pie(df_hour.groupby('season')['cnt'].mean().values, labels=['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'], autopct='%1.1f%%')
plt.title('Persentase Rata-rata Jumlah Pengguna Sepeda Berdasarkan Musim (Jam)')
plt.show()

# Visualisasi data 4: Persentase rata-rata jumlah pengguna sepeda berdasarkan musim (hari)
plt.figure(figsize=(10,6))
plt.pie(df_day.groupby('season')['cnt'].mean().values, labels=['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'], autopct='%1.1f%%')
plt.title('Persentase Rata-rata Jumlah Pengguna Sepeda Berdasarkan Musim (Hari)')
plt.show()
```

**Insight:**
- Visualisasi data menunjukkan bahwa cuaca memiliki pengaruh yang signifikan terhadap penggunaan sepeda berbagi, dan bahwa ada perbedaan penggunaan sepeda berbagi antara hari kerja dan hari libur.
